#Problem 2

 Load Metro_Interstate_Traffic_Volume.csv as DataFrame

● From ‘date_time’ column:

○ Create ‘month’ column (10 points)

○ Create ‘day_of_month’ column (digit tanggal) (10 points)

○ Create ‘day_name’ column (Monday, Tuesday, ...) (10 points)

○ Create ‘hour’ column (digit jam) (10 points)

○ Create ‘is_weekend’ binary column (1 if day == saturday OR sunday, 0
otherwise) (10 points)



```python
#Load Metro_Interstate_Traffic_Volume.csv as DataFrame

df2=pd.read_csv('//content/drive/MyDrive/dibimbing/Day 12 Python data manipulation/Metro_Interstate_Traffic_Volume.csv')
df2
```

```python
#Create ‘month’ column (10 points)

#ensure date time format in the dataset is pandas standard format
df2['date_time']=pd.to_datetime(df2['date_time'])
df2
```

```python
#adding the colmn month with value stripped form column date_time

df2['month']=df2['date_time'].dt.month
df2.head()
```

```python
#Create ‘day_of_month’ column (digit tanggal) (10 points)

df2['day_of_month']=df2['date_time'].dt.day
df2.head()
```

```python
df2.dtypes
```

```python
#Create ‘day_name’ column (Monday, Tuesday, ...) (10 points)

df2['day_name']=df2['date_time'].dt.day_name()
df2.head()
```

```python
#Create ‘hour’ column (digit jam) (10 points)

df2['hour']=df2['date_time'].dt.hour
df2.head()
```

```python
#Create ‘is_weekend’ binary column (1 if day == saturday OR sunday, 0 otherwise) (10 points)

df2['is_weekend']=df2['day_name'].apply(lambda x: 1 if x == 'Sunday' or x=='Saturday' else 0)

df2
```

```python
#to check is_weekend column correct

df2.groupby('day_name').mean()
```
